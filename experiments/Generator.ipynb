```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.options.display.max_rows = 100
pd.options.display.max_columns = 50
```

## ScikitMultiflow stream

```python
from skmultiflow.data.waveform_generator import WaveformGenerator
```

```python
stream = WaveformGenerator(has_noise = False)
stream.prepare_for_use()
```

```python
data = stream.next_sample(2000)
```

```python
df1 = pd.DataFrame(data[0])
df2 = pd.DataFrame(data[1])
df2 = df2.rename({0:'class'}, axis = 1);
```

```python
df = pd.concat([df2,df1], axis = 1)
```

```python
df
```

```python
df.describe()
```

```python
fig, ax = plt.subplots(); 
fig.set_size_inches(10, 5)
ax.plot(range(2000), df['class'],  '-');

plt.show();
```

```python
for i in range(df.shape[1] - 1):
    fig, ax = plt.subplots(); 
    fig.set_size_inches(10, 5)
    ax.plot(range(2000), df[i],  '-');
#    ax.legend(loc=1, borderaxespad=1)
#    plt.xlabel('')
#    plt.ylabel('')
    plt.show();
```

```python
for i in range(df.shape[1] - 1):
    plt.hist(df[i], 15, alpha=0.5, density=False)
    plt.show()
```
